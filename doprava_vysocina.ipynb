```python
def vysledny_cas(cesta):
    vysledek = 0
    for section in cesta['routes'][0]['sections']:
        vysledek += section['travelSummary']['duration']
    return vysledek
```

```python
for i in vysocina.index:
    if i < 410:
        try:
            start_lat = vysocina['Zeměpisná šířka'][i]
            start_lon = vysocina['Zeměpisná délka'][i]
            cil_lat = vysocina['latitude'][i]
            cil_lon = vysocina['longitude'][i]
            adresa = f"https://router.hereapi.com/v8/routes?transportMode=car&origin={start_lat},{start_lon}&destination={cil_lat},{cil_lon}&return=summary&apikey=BQbRWch3xMjNI53J7ye8FPvHnqamqQnJyiZnneG4JSQ&arrivalTime=2022-11-16T08:00:00"
            vysledek = requests.get(adresa)
            cesta = vysledek.json()
            cas = cesta["routes"][0]["sections"][0]["summary"]["duration"]
            vysocina['cas_cesty'][i] = cas
        except IndexError:
            vysocina['cas_cesty'][i] = 'NaN'
```
