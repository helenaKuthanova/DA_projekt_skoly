```python
import pandas

file_obce = 'sldb2021_vek5_pohlavi.csv'
df_obce = pandas.read_csv(file_obce, encoding='utf-8')
df_obce
```

## Filtruji věkové kategorie

```python
unikatni_vek_txt = df_obce['vek_txt'].unique()
print(unikatni_vek_txt, len(unikatni_vek_txt))
```

## Můžu smazat všechny sloupce pohlaví a věků

```python
df_obce_teenageri_bezpohlavni = df_obce_teenageri_bezpohlavni.drop(['pohlavi_kod', 'vek_cis', 'vek_kod', 'pohlavi_cis', 'pohlavi_txt', 'sldb_datum', 'vek_txt'], axis=1)
df_obce_teenageri_bezpohlavni
```

## Filtruji hodnoty pro kraje

```python
unikatni_uzemi_cis = df_obce['uzemi_cis'].unique()
print(unikatni_uzemi_cis, len(unikatni_uzemi_cis))
```

##### Číselník pro kraj: 100, okres: 101, správní obvody obcí s rozšířenou působností: 65, správní obvody v hlavním městě: 72

```python
df_kraje = df_obce_teenageri_bezpohlavni[df_obce_teenageri_bezpohlavni['uzemi_cis'] == 100]
df_kraje
```

## Ukládám tabulku podle krajů

```python
df_kraje = df_kraje.reset_index(drop=True)  # reset index, drop původní index, PK je IDHOD + UZEMI_KOD
df_kraje
```

```python
df_kraje.to_csv('teenageri_kraje.csv')
```

## Filtruji hodnoty pro okresy a nahrazuju Prahu za 22 městských obvodů
ČR má 76 okresů (#101), Praha není okres, Praha má 22 správních obvodů (P1-22, #72)

```python
df_okresy = df_obce_teenageri_bezpohlavni[df_obce_teenageri_bezpohlavni['uzemi_cis'] == 101]
df_okresy
```

```python
df_okresy = df_okresy[df_okresy['uzemi_kod'] != 40924]  # 40924 = uzemi_kod pro Prahu
df_okresy
```

```python
df_praha = df_obce_teenageri_bezpohlavni[df_obce_teenageri_bezpohlavni['uzemi_cis'] == 72]
df_praha
```

```python
df_okresy_praha = pandas.concat([df_okresy, df_praha])  # kontrola: 76 + 22 = 98
df_okresy_praha
```

### Ukládám finální tabulku okresů + pražských správních obvodů
(= 98 ř.)

```python
df_okresy_praha = df_okresy_praha.reset_index(drop=True)  # resetuju index
df_okresy_praha
```

```python
df_okresy_praha.to_csv('teenageri_okresy.csv')
```
