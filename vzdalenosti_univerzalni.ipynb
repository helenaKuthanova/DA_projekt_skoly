### kategorie olomouc, 0-2000, 2000-10.000, 10.000-50.000

```python
O2000 = kraj[kraj['celkem'] < 2000]
len(O2000), O2000['celkem'].sum()
```

```python
O10000 = kraj[(kraj['celkem'] < 10000) & (kraj['celkem'] >= 2000)]
len(O10000), O10000['celkem'].sum()
```

```python
O50000 = kraj[(kraj['celkem'] >= 10000) & (kraj['celkem'] < 50000)]
len(O50000), O50000['celkem'].sum()
```

```python
sample_O2000 = O2000.sample(10)
sample_O2000
```

```python
sample_O10000 = O10000.sample(10)
sample_O10000
```

```python
sample_O50000 = O50000
O50000
```

```python
sample_O = pandas.concat([sample_O2000, sample_O10000, sample_O50000])
sample_O
```

```python
kebo = sample_O['Název obce']
kebo
```

```python
kebo.to_csv('kebo.csv', index=False)
```

```python
# HRADEK, sedlec, rozstani, sloup, osek
vzdalenosti_hradek = pandas.DataFrame()
for i in skoly.index:
    skola = (skoly['Zeměpisná šířka'][i], skoly['Zeměpisná délka'][i])
    vzdalenost = haversine(hradek, skola)
    if vzdalenost < 30:
        radek = skoly.iloc[i] 
        radek['Hradek'] = round(vzdalenost, 0)
        vzdalenosti_hradek = vzdalenosti_hradek.append(radek)
vzdalenosti_hradek  # 31 r.
```
