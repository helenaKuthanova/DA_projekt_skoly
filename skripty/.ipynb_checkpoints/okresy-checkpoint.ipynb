```python
import pandas

file_obce = 'sldb2021_vek5_pohlavi.csv'
df_obce = pandas.read_csv(file_obce, encoding='utf-8')
df_obce
```

## Filtruji věkové kategorie
##### Pro kontrolu hledám unikátní hodnoty a jejich počet v některých sloupcích

```python
unikatni_vek_kod = df_obce['vek_kod'].unique()
print(unikatni_vek_kod, len(unikatni_vek_kod))
```

```python
unikatni_vek_txt = df_obce['vek_txt'].unique()
print(unikatni_vek_txt, len(unikatni_vek_txt))
```

## Filtruji hodnoty pro kraje

```python
unikatni_uzemi_cis = df_obce['uzemi_cis'].unique()
print(unikatni_uzemi_cis, len(unikatni_uzemi_cis))
```

##### Číselník pro kraj: 100, okres: 101, správní obvody obcí s rozšířenou působností: 65, správní obvody v hlavním městě: 72, městský obvod, městská část: 44, obec: 43

```python
df_kraje = df_obce_teenageri_bezpohlavni[df_obce_teenageri_bezpohlavni['uzemi_cis'] == 100]
df_kraje
```

## Ukládám tabulku podle krajů

```python
df_kraje = df_kraje.reset_index(drop=True)  # reset index, drop původní index, PK je IDHOD + UZEMI_KOD
df_kraje
```

```python
df_kraje.to_csv('teenageri_kraje.csv')
```

## Filtruji hodnoty pro okresy a nahrazuju Prahu za ???
ČR má 76 okresů (#101), Praha není okres, Praha má 22 správních obvodů (P1-22, #72) => 10 městských obvodů

```python
df_okresy = df_obce_teenageri_bezpohlavni[df_obce_teenageri_bezpohlavni['uzemi_cis'] == 101]
df_okresy
```

```python
df_okresy = df_okresy[df_okresy['uzemi_kod'] != 40924]  # 40924 = uzemi_kod pro Prahu
df_okresy
```

```python
df_praha = df_obce_teenageri_bezpohlavni[df_obce_teenageri_bezpohlavni['uzemi_cis'] == 72]
df_praha
```

## Sčítám MČ do 10 obvodů
11=4, 12=4, 13=5, 14=9, 15=10, 16=5, 17=6, 18-21 = 9, 22=10

 ###### Praha je rozdělena do 10 obvodů, které jsou územními jednotkami na úrovni okresů.
 ###### Počty obyvatel při dělení na 22 částí jsou nízké relativně k okresům.

```python
df_pracovni_praha = df_praha
df_pracovni_praha.insert(8, 'nazev', ['Praha 1', 'Praha 2', 'Praha 3', 'Praha 4', 'Praha 5', 'Praha 6', 'Praha 7', 'Praha 8', 'Praha 9', 'Praha 10', 'Praha 4', 'Praha 4', 'Praha 5', 'Praha 9', 'Praha 10', 'Praha 5', 'Praha 6', 'Praha 9', 'Praha 9', 'Praha 9', 'Praha 9', 'Praha 10'], allow_duplicates=True)
```

```python
df_pracovni_praha = df_pracovni_praha.iloc[:, :-1]  # odstraňuju sloupce 'nazev', když je jich moc
df_pracovni_praha
```

```python
df_pracovni_praha = df_pracovni_praha.groupby(by='nazev').sum(['hodnota'])
df_pracovni_praha
```

```python
df_pracovni_praha = df_pracovni_praha.drop(['idhod', 'ukaz_kod', 'uzemi_cis', 'uzemi_kod', 'sldb_rok'], axis=1)
df_pracovni_praha
```

```python
df_pracovni_praha.reset_index()
```

```python
df_pracovni_okresy = df_okresy
df_pracovni_okresy['nazev'] = df_pracovni_okresy['uzemi_txt'] 
```

```python
df_pracovni_okresy = df_pracovni_okresy.drop(['idhod', 'ukaz_kod', 'uzemi_cis', 'uzemi_kod', 'sldb_rok', 'ukaz_txt', 'uzemi_txt'], axis=1)
```

```python
df_pracovni_okresy = df_pracovni_okresy[['nazev', 'hodnota']]
df_pracovni_okresy
```

## Ukládám finální tabulku okresů + 10 pražských obvodů
(= 86 ř.)

```python
df_pracovni = pandas.concat([df_pracovni_okresy, df_pracovni_praha])  # 76 + 10
df_pracovni
```

```python
df_pracovni = df_pracovni.reset_index(drop=True)
df_pracovni
```

```python
df_pracovni.to_csv('teenageri_okresy10.csv')
```

## [ Spojuju okresy s 22 MČ ]
