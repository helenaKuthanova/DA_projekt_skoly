```python
import pandas as pd
skoly_vsechny_kraje_praha = pd.read_csv("skoly_vsechny_kraje.csv", encoding='utf-8')
```

```python
skoly_vsechny_kraje_praha.loc[skoly_vsechny_kraje_praha['kraj'] == 'Hlavní město Praha', "okres"] = "Praha"
```

```python
ciselnik_okresy = pd.read_csv("CIS0109_CS1.csv", encoding = "utf-8", sep=",")
ciselnik_okresy.head(5)
```

```python
ciselnik_okresy = ciselnik_okresy.drop(columns = ["KODJAZ", "AKRCIS", "KODCIS", "ZKRTEXT", "ADMPLOD", "ADMNEPO"])
ciselnik_okresy = ciselnik_okresy.rename(columns = {"CHODNOTA":"kod_okresu_a", "TEXT":"okres", "KODRSO": "kod_okresu_b"})
```

```python
ciselnik_okresy.info()
```

```python
skoly_vsechny_kraje_praha = pd.merge(skoly_vsechny_kraje_praha, ciselnik_okresy, how="left", on="okres")
```

```python
skoly_vsechny_kraje_praha.info()
```

```python
skoly_vsechny_kraje_praha.to_csv("skoly_vsechny_kraje_praha.csv")
```
