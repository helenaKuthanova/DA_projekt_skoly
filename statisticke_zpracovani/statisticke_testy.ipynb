```python
res = stats.kruskal(olomoucky_cas, vysocina_cas, stredocesky_cas, jihocesky_cas, jihomoravsky_cas, karlovarsky_cas, kralovehradecky_cas, liberecky_cas, moravskoslezsky_cas, pardubicky_cas, plzensky_cas, ustecky_cas, zlinsky_cas)
print(res)

```

## Test korelace: Spearmanův koeficient
Zjišťujeme, jestli existuje závislost mezi počtem obyvatel a časem cesty.

```python
import seaborn
```

```python
res = stats.spearmanr(data["celkem"], data["cas_cesty"])
print(res)
```

Velmi slabá (téměř nulová) korelace mezi veličinami, statisticky ale signifikantní.
To znamená, že díky velkému množství dat je výsledek výpočtu velmi přesně odpovídající distribuci v populaci, ale protože je korelace extrémně slabá, nemá téměř žádný vliv na délku cesty do školy.

```python
g = seaborn.JointGrid(data=data, x="celkem", y="cas_cesty")
g.plot_joint(seaborn.scatterplot, legend=False)
plt.show()
```

```python
#medián za obec všechna data:
median_cas_na_obec = data.groupby(["obec_bydliste", "celkem"]).aggregate({'cas_cesty':'median'})


g = seaborn.JointGrid(data=median_cas_na_obec, x="celkem", y="cas_cesty")
g.plot_joint(seaborn.scatterplot, legend=False)
plt.show()


```

```python
median_cas_na_obec = median_cas_na_obec.reset_index(names=['obec_bydliste', 'celkem'])
median_cas_na_obec
```

```python
# Jaký vliv má použití mediánu na výpočet korelace? 
res = stats.spearmanr(median_cas_na_obec["celkem"], median_cas_na_obec["cas_cesty"])
print(res)  
```

i z hlediska mediánu je to velmi slabá korelace, medián stejně jako práce s plnými daty odráží hustotu osídlení v ČR
